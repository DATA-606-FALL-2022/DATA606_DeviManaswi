## Finding the Null values

## Replacing the Null values

```python
applicant_records['OCCUPATION_TYPE']=applicant_records['OCCUPATION_TYPE'].fillna(0).inpalce=True
```

```python
applicant_records.isnull().sum()
```

```python
applicant_records.columns
```

```python
credit_records.isnull().sum()
```

### 'STATUS' column is the potential feature having predictive power for machine learning

```python
credit_records['STATUS'].unique()
```

```python
credit_records.STATUS.value_counts()
```

```python
credit_records['STATUS'] = np.where((credit_records['STATUS'] == '2') | (credit_records['STATUS'] == '3' )| (credit_records['STATUS'] == '4' )| (credit_records['STATUS'] == '5'), 1, 0)
```

```python
credit_records.STATUS.value_counts()
```

## Merging two datasets

```python
merged_df=pd.merge(applicant_records,credit_records,on='ID')
```

```python
merged_df
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

## Correlation between the columns

```python
merged_df.corr()
```

## Representation of correlation using heatmap

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(merged_df.corr(), cmap="YlGnBu");
```

### Clearly we can see that there is no relation between occupation type and flag_mobil with rest of the columns. So we can drop these columns

```python
final_df=merged_df.drop(['FLAG_MOBIL', 'OCCUPATION_TYPE'], axis=1)
```

```python
final_df.info()
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(final_df.corr(), cmap="YlGnBu");
```

### Visualizing distribution of target variable

```python
plt.figure(figsize=(10,10))
sns.countplot(x="STATUS", data=final_df, palette="YlGnBu")
plt.show()
```
